<p>2. Find the top 5 countries for coffee for the year 2013. </p> 
<ol style = "list-style-type: lower-alpha">
    <li>Check the CSV for how coffee is labeled</li>
    <li>Group the top 5 countries for 2013</li> 
    <li>Build the visualization with Seaborn</li>
</ol>

```python
df_first_specific = df[df.Item == 'Coffee and products'].copy()
df_first_specific.loc[:, 'Percentage'] = df_first_specific.Y2013 / \
df_first_specific.Y2013.sum()
df_first_specific = df_first_specific.sort_values(by = 'Y2013', ascending = False)
df_first_specific.loc[:, 'Cumsum'] = df_first_specific.Percentage.cumsum()
df_first_specific[['Area', 'Y2013', 'Percentage', 'Cumsum']]
```

```python
top_coffee = df_first_specific.nlargest(5, 'Y2013')
```

```python
sns.set(rc = {'figure.figsize': (12,12)})
plt.xticks(rotation = 70)
sns.barplot(x = 'Area', y = 'Y2013', data = top_coffee, \
           label = 'item')
```

<p>3. Compare the top 5 countries for coffee from the year 2012 to 2013.</p> 
<ol style = "list-style-type: lower-alpha">

```python
top_coffee_2012 = df_first_specific.nlargest(5, 'Y2012')
top_coffee_2012.head(5)
```

```python
sns.set(font_scale = 1.5)
plt.xticks(rotation = 70)
sns.set_color_codes('pastel')

sns.barplot(x = "Area", y = 'Y2013', data = top_coffee, label = '2013', \
           color = 'r')
sns.barplot(x = "Area", y = 'Y2012', data = top_coffee_2012, \
            label = '2012', color = 'b')

ax = plt.gca()
ax.legend(ncol = 2, loc = 'lower right', frameon = True)
ax.set(xlim = (0, 6), ylabel = "Coffee", xlabel = "Country")
sns.despine(left = True, bottom = True)
```
